# Notebook to convert properties.csv as provided by BeCode into a useable dataset

```python
import pandas as pd

#____loading data____
df = pd.read_csv('../data/properties.csv')

#____selecting only houses____
df_house = df[df['property_type'] == 'HOUSE']

#____selecting only apartments____
df_app = df[df['property_type'] == 'APARTMENT']
```

```python
import pandas as pd

# ____alligning columnnames____
new_column_names = {
    'id': 'property_id',
    'price': 'price',
    'subproperty_type': 'property_subtype',
    'province': 'province',
    'locality': 'locality_name',
    'zip_code': 'postal_code',
    'latitude': 'latitude',
    'longitude': 'longitude',
    'total_area_sqm': 'living_area',
    'surface_land_sqm':'surface_of_good',
    'nbr_frontages': 'number_of_facades',
    'equipped_kitchen': 'kitchen_type',
    'fl_furnished': 'furnished',
    'fl_open_fire': 'open_fire',
    'fl_terrace': 'terrace',
    'terrace_sqm': 'terrace_area',
    'fl_garden': 'garden',
    'garden_sqm': 'garden_area',
    'fl_swimming_pool': 'swimming_pool',
    'state_building': 'state_of_building'    
}

df_house.rename(columns=new_column_names, inplace=True)
df_app.rename(columns=new_column_names, inplace=True)

# ______Drop the columns that are not included in the new list______
columns_to_drop = [col for col in df_house.columns if col not in new_column_names.values()]
df_house.drop(columns=columns_to_drop, inplace=True)

# ______Drop the columns that are not included in the new list______
columns_to_drop = [col for col in df_app.columns if col not in new_column_names.values()]
df_app.drop(columns=columns_to_drop, inplace=True)
```
